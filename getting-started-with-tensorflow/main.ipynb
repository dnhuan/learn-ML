```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create noise
x = np.linspace(0, 1, 1000) # array from 0 to 1, 1000 steps
y = np.random.randn(1000) * 0.3 + 2 * x # randn - normal distribution, rand - uniform
plt.figure(figsize=(10, 8))
plt.scatter(x, y, s=2) # dot size = 2
plt.xlabel('x (features)')
plt.ylabel('y (target)')
plt.show()
```
